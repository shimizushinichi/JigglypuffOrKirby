```
from tensorflow.keras import models
from tensorflow.keras import optimizers
from keras.utils import np_utils
import numpy as np
import os
import warnings
warnings.simplefilter("ignore")

# 評価用データを読み込み
###TODO_testdata_converterで出力したファイル名をeval_numpyに設定###
eval_numpy = "sample_npy"
eval_X, eval_Y = np.load("/content/drive/My Drive/Colab Notebooks/"+eval_numpy, allow_pickle=True)
eval_Y = np_utils.to_categorical(eval_Y, 2)
```

```
#転移学習で出力されたモデルの評価を行う
input_dir = "/content/drive/My Drive/Colab Notebooks/transfer_learning/"
filelist = os.listdir(input_dir)
for filename in filelist:
  model = models.load_model(input_dir + filename)
  model.compile(loss="categorical_crossentropy",
             optimizer=optimizers.RMSprop(lr=1e-5),
             metrics=["acc"])
  score = model.evaluate(x=eval_X, y=eval_Y)
  print(filename)
  print("loss=", score[0])
  print("accuracy=", score[1])
```

```
#ファインチューニングで出力されたモデルの評価を行う
input_dir = "/content/drive/My Drive/Colab Notebooks/model/finetuning/"
filelist = os.listdir(input_dir)
for filename in filelist:
  model = models.load_model(input_dir + filename)
  model.compile(loss="categorical_crossentropy",
             optimizer=optimizers.RMSprop(lr=1e-5),
             metrics=["acc"])
  score = model.evaluate(x=eval_X, y=eval_Y)
  print(filename)
  print("loss=", score[0])
  print("accuracy=", score[1])
```
